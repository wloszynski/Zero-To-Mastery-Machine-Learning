## Matplotliv example workflow

```python
# 0. import matplotliv and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Preapre data
x = [1, 220, 1, 4, 100]
y = [11, 0, 323, -29, 100]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("./sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond  butter": 10,
                    "Peanut butter": 8, 
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("7/17/2020",
               periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
```

```python
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("7/17/2020", periods=len(car_sales))
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales["Total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sale Date", y="Total sales")
```
